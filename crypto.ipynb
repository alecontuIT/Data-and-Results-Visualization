```python
import pandas as pd
import numpy as np

# BITCOIN
bit = pd.read_csv('data/crypto/bitcoin.csv')
bit['Date'] = pd.to_datetime(bit["Date"])
bit = bit.set_index(['Date'])
bit = bit.asfreq("D", method="ffill")
bit = bit[bit.index >= pd.Timestamp("2018-01-01")]

# CARDANO
cardano = pd.read_csv('data/crypto/cardano.csv')
cardano['Date'] = pd.to_datetime(cardano["Date"])
cardano = cardano.set_index(['Date'])
cardano = cardano.asfreq("D", method="ffill")

# DASH
dash = pd.read_csv('data/crypto/dash.csv')
dash['Date'] = pd.to_datetime(dash["Date"])
dash = dash.set_index(['Date'])
dash = dash.asfreq("D", method="ffill")

# ETHEREUM
eth = pd.read_csv('data/crypto/ethereum.csv')
eth['Date'] = pd.to_datetime(eth["Date"])
eth = eth.set_index(['Date'])
eth = eth.asfreq("D", method="ffill")
eth = eth[eth.index >= pd.Timestamp("2018-01-01")]

# LITECOIN
lite = pd.read_csv('data/crypto/litecoin.csv')
lite['Date'] = pd.to_datetime(lite["Date"])
lite = lite.set_index(['Date'])
lite = lite.asfreq("D", method="ffill")

```

```python
crypto = pd.DataFrame()  
crypto.index= bit.index
crypto["Bitcoin"] = bit.Close
crypto["Cardano"] = cardano.Close
crypto["Dash"] = dash.Close
crypto["Ethereum"] = eth.Close
crypto["Litecoin"] = lite.Close

#crypto = (crypto - crypto.min()) / (crypto.max() - crypto.min())
```

```python
crypto = pd.DataFrame()  
crypto.index= bit.index
crypto["Bitcoin"] = bit.Volume
crypto["Cardano"] = cardano.Volume
crypto["Dash"] = dash.Volume
crypto["Ethereum"] = eth.Volume
crypto["Litecoin"] = lite.Volume
crypto = crypto.asfreq("M")
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Bar(x = crypto.index, y = crypto.Dash, name = "Dash"))
fig.add_trace(go.Bar(x = crypto.index, y = crypto.Litecoin, name = "Litecoin"))
fig.add_trace(go.Bar(x = crypto.index, y = crypto.Cardano, name = "Cardano"))
fig.add_trace(go.Bar(x = crypto.index, y = crypto.Ethereum, name = "Ethereum"))
fig.add_trace(go.Bar(x = crypto.index, y = crypto.Bitcoin, name = "Bitcoin"))


fig.update_layout(
    updatemenus=[
        dict(
            type = "buttons",
            direction = "left",
            buttons=list([
                dict(
                    args=["type", "bar"],
                    label="Barplot",
                    method="restyle"
                ),
                dict(
                    args=[{"type":"scatter","stackgroup":"one",}],
                    label="Scatter",
                    method="restyle"
                )
            ]),
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.11,
            xanchor="left",
            y=1.1,
            yanchor="top"
        ),
    ]
)

# Add range slider
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            # Colore dei bottoni quando ci passi sopra con mosue
            #activecolor = "red",
            # Colore dei bottoni
            #bgcolor = "yellow",
            #Posizione dei bottoni sull'asse x (valori tra -2 e 3)
            x = 0.5,
            buttons=list([
                dict(count=6,
                     label="6 months",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="1 year",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

fig.update_layout(
    # Modifica titolo
    title={
        'text': "Cryptocurrencies' volume traded montly",
        'y':0.95,
        'x':0.5,
        "font.size": 20,
        'xanchor': 'center',
        'yanchor': 'top'}
)


fig.update_layout(barmode = "stack",  xaxis_tickangle=-45)

fig.show()

#fig.write_html("./grafici/crypto_volume_trade.html")
```

```python
crypto = pd.DataFrame()  
crypto.index= bit.index
crypto["Bitcoin"] = bit.Volume
crypto["Cardano"] = cardano.Volume
crypto["Dash"] = dash.Volume
crypto["Ethereum"] = eth.Volume
crypto["Litecoin"] = lite.Volume
crypto = crypto.asfreq("Y")
crypto.index = pd.to_datetime(crypto.index, format='%Y-%m-%d').year
# Transpose matrix
crypto = crypto.T
```

```python
crypto["Colors"] = ["FFA15A", "00CC96","636EFA", "AB63FA", "EF553B"]
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

labels = crypto.index
specs = [[{'type':'domain'}, {'type':'domain'}], [{'type':'domain'}, {'type':'domain'}]]
fig = make_subplots(rows=2, cols=2, specs = specs, subplot_titles=["2018","2019","2020","2021"])
fig.add_trace(go.Pie(labels=labels, values=crypto[2018], name='2018'), 1,1)
fig.add_trace(go.Pie(labels=labels,  values=crypto[2019], name='2019'), 1,2)
fig.add_trace(go.Pie(labels=labels,  values=crypto[2020], name='2020'), 2,1)
fig.add_trace(go.Pie( labels=labels, values=crypto[2021], name='2021'), 2,2)

fig.update_traces(marker=dict(colors=crypto.Colors))

fig.update_layout(margin = dict(t=100, l=0, r=0, b=0))
fig = go.Figure(fig)

fig.update_layout(
    # Modifica titolo
    title={
        'text': "Cryptocurrencies' volume traded yearly",
        'y':0.95,
        'x':0.5,
        "font.size": 30,
        'xanchor': 'center',
        'yanchor': 'top'}
)

fig.update_layout(legend=dict(
    yanchor="middle",
    y=0.5,
    xanchor="center",
    x=0.5,
    borderwidth = 1,
    bordercolor = "royalblue",
    bgcolor = "#e6f9ff"
))

fig.update_traces(texttemplate='%{value:$.2s}',textposition='auto' )

fig.add_annotation(text="* M = Million \n G = Billion",
                  xref="paper", yref="paper",
                  x=0.5, y=0.3, showarrow=False)
fig.show()
#fig.write_html("./grafici/crypto_trade_year.html")
```
